```python
from time import sleep
from pynq import Overlay
from pynq import GPIO

ps_gpio_design = Overlay("./ps_gpio_kv260.bit")

short_delay = 0.5
delay = 2
counter_delay = 9



green_traffic = GPIO(GPIO.get_gpio_pin(0), 'out')   # initialize green traffic light pin as input
yellow_traffic = GPIO(GPIO.get_gpio_pin(1), 'out')         # initialize LED pin as output
red_traffic = GPIO(GPIO.get_gpio_pin(2), 'out')  # initialize slider LED pin as output
green_ped = GPIO(GPIO.get_gpio_pin(3), 'out')   # initialize green traffic light pin as input
yellow_ped = GPIO(GPIO.get_gpio_pin(4), 'out')  # initialize slider LED pin as output
red_ped = GPIO(GPIO.get_gpio_pin(5), 'out')  # initialize slider LED pin as output
switch = GPIO(GPIO.get_gpio_pin(6), 'in')  # initialize slider LED pin as output
counter = GPIO(GPIO.get_gpio_pin(7), 'out')  # initialize slider LED pin as output

green_traffic.write(0)
counter.write(0)
yellow_traffic.write(0)
red_traffic.write(0)
green_ped.write(0)
red_ped.write(0)
yellow_ped.write(0)

while True:
    slider = switch.read()
    S1()
    
    if slider > 0:
        S2()
        S3()
    else:
        S1()

def S1():
    green_traffic.write(1)
    counter.write(0)
    yellow_traffic.write(0)
    red_traffic.write(0)
    green_ped.write(0)
    red_ped.write(1)
    yellow_ped.write(0)
    
    return 

def S2():
    green_traffic.write(0)
    sleep(0.5)

    red_ped.write(1)
    sleep(0.5)

    yellow_traffic.write(1)
    sleep(5)
    yellow_traffic.write(0)
    sleep(0.5)

    red_traffic.write(1)
    red_ped.write(0)
    sleep(0.5)
    
def S3():
    green_ped.write(1)
    counter.write(1)

    red_ped.write(0)
    #counter.write(1)
    sleep(10)
    green_ped.write(0)
    sleep(0.5)
    yellow_ped.write(1)
    sleep(5)
    yellow_ped.write(0)
    sleep(0.5)

    red_ped.write(1)
    counter.write(0)
    sleep(0.5)
    
```
